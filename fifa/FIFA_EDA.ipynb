```python
def convertValue(value):
    output = value.replace('€', '')
    if 'M' in output:
        output = float(output.replace('M', ''))*1000000
    elif 'K' in value:
        output = float(output.replace('K', ''))*1000
    return float(output)
```

```python
fifa["Value"] = fifa["Value"].apply(lambda x: convertValue(x))
```

```python
fifa["Value"].head()
```

```python
def convertValue(value):
    output = value.replace('€', '')
    if 'M' in output:
        output = float(output.replace('M', ''))*1000000
    elif 'K' in value:
        output = float(output.replace('K', ''))*1000
    return float(output)
```

```python
fifa["Wage"] = fifa["Wage"].apply(lambda x: convertValue(x))
```

```python
fifa["Wage"]
```

```python
fifa['Height']=fifa['Height'].astype(str)
fifa['Weight']=fifa['Weight'].astype(str)
fifa.Height=[x.replace("'",'.') for x in fifa.Height]
fifa.Weight=[x.replace("lbs",'') for x in fifa.Weight]
```

```python
fifa['Height']=fifa['Height'].astype(float)
fifa['Weight']=fifa['Weight'].astype(float)
```

```python
fifa['Release Clause']=fifa['Release Clause'].astype(str)
fifa['Release Clause']=[x.strip('€') for x in fifa['Release Clause']]
fifa['Release Clause']=[x.replace('M','0000000') for x in fifa['Release Clause']]
fifa['Release Clause']=[x.replace('K','000') for x in fifa['Release Clause']]
fifa['Release Clause']=fifa['Release Clause'].astype(float)
```

```python
fifa['Contract Valid Until']=pd.to_datetime(fifa['Contract Valid Until'])
```

```python
# 4.Generate tables containing the top 5 players by Overall rating for each unique position.
print("Following are %d unique positions: " %fifa['Position'].nunique()); print()
tmp=fifa['Position'].value_counts()
print(tmp)
```

```python
# Creating Dictionary with each key as postion and value as Dataframe with top 5 players in that position
Dict1=dict()
df=pd.DataFrame()
for index in tmp.index:
    df=fifa[fifa['Position']==index].sort_values(by="Overall", ascending=False).head(5)
    Dict1[index]=df
```

```python
# Print the top5 players names (printing entire df makes output messy) in each position.
for i,k in Dict1.items():
    print("Position: ",i)
    print(k.Name)
    print("***")
    print()
```

```python
# 4. a)	Are there any players appearing in more than one Table. Please point out such players

# Creating a list of all top 5 players in each position
Names=[]
for i,k in Dict1.items():
    Names.extend(list(k.Name))
Names=[x.lower() for x in Names]
print(Names)
```

```python
# b)What is the average wage one can expect to pay for the top 5 in every position?
for i,k in Dict1.items():
    print("Average Wage for Position:",i, end=" ")
    print("is ",k.Wage.mean())
    print()
```
